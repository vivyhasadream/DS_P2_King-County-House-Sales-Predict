```python
avg_sqft_grade = df.groupby("grade_str")["sqft_living"].mean().sort_values()
avg_price_grade = df.groupby("grade_str")["price"].mean().sort_values()
```

```python
import seaborn as sns
import matplotlib.ticker as ticker
fig, ax = plt.subplots(figsize=(35,20))
ax.bar(avg_sqft_grade.keys(), avg_sqft_grade, color = "rgbgymc")
plt.rcParams.update({'font.size':45})
ax.set_xlabel('Level of Grade')
ax.set_ylabel("sqft_living")
ax.set_title('sqft_living based on grades ')
ax.set_xticklabels(avg_sqft_grade.keys(), rotation=30)
ax.get_yaxis().set_major_formatter(
    ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.grid()
plt.tight_layout()3
plt.show();
```

```python
#lets check the distribution of bedrooms over the price
fig, ax = plt.subplots(figsize=(35, 15))
#plt.rcdefaults()
# sets up bar plot
plt.rcParams.update({'font.size':35})
ax.barh(avg_price_grade.keys(), avg_price_grade, color = "rgbgymc")
ax.set_xlabel('Prices of houses(K)')
ax.set_ylabel("Grades")
ax.set_title('Average price of houses based on grades')
#ax.set_xticklabels(director_movie_profit.keys(), rotation=45)
#ax.get_yaxis().set_major_formatter(
    #ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
#plt.rcParams.update({'font.size': 10})
plt.grid()
plt.tight_layout()
plt.show();
```
